```python
import os
import re
```

```python
fnames_gb = os.listdir('dataset_GB')
tickers_gb = [fname.split('-')[0] for fname in fnames_gb]
tickers_gb = list(set(tickers_gb))
tickers_gb
```

```python
fnames_vb = os.listdir('dataset_VB')
tickers_vb = [fname.split('-')[0] for fname in fnames_vb]
tickers_vb = list(set(tickers_vb))
tickers_vb
```

```python
overlap_tickers = []
for ticker_gb in tickers_gb:
    if ticker_gb in tickers_vb:
        overlap_tickers.append(ticker_gb)
```

```python
overlap_tickers  # tickers that exist both in GB and VB; therefore, these tickers should be removed from the VB list.
```

```python
len(overlap_tickers)
```

### Remove the overlaopped tickers 

```python
fnames_vb = os.listdir('dataset_VB')
```

```python
count = 0
for overlap_ticker in overlap_tickers:
    for fname_vb in fnames_vb:
        if re.match(f'{overlap_ticker}-.*?', fname_vb):
            count += 1
            os.unlink(f'dataset_VB/{fname_vb}')
```

# Analyze VB-info

```python
import os
import pandas as pd
```

```python
info_fnames_vb = os.listdir('dataset_VB')
info_fnames_vb = [fname for fname in info_fnames_vb if '-info' in fname]
```

```python
df_info_vb = pd.DataFrame()
for fname in info_fnames_vb:
    ticker = fname.split('-')[0]
    df_ = pd.read_csv(f'dataset_VB/{fname}')
    
    if df_.shape[0] > 1:
        df_ = df_.iloc[:1, :]
    
    df_.index = [ticker]
    df_info_vb = pd.concat((df_info_vb, df_))
```

```python
df_info_vb
```

```python
df_info_vb.to_csv('dataset_VB_info.csv')
```

```python
df_info_vb.isna()
```

# remove columns with lots of nans above threshold

```python
threshold = 0.3

valid_cols = []
for colname in df_info_vb.columns:
    na_ratio = df_info_vb.isna()[colname].mean()
    print(colname, na_ratio)
    
    if na_ratio <= threshold:
        valid_cols.append(colname)
```

```python
valid_cols
```

```python
clean_df_info_vb = df_info_vb[valid_cols]
```

# remove rows with nans 

```python
clean_df_info_vb = clean_df_info_vb.dropna()
```

```python
clean_df_info_vb.to_csv('dataset_VB_info_clean.csv')
```
